```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
housing = fetch_california_housing()

# Create a DataFrame with the dataset
house_df = pd.DataFrame(housing.data, columns=housing.feature_names)
house_df['Price'] = housing.target
# Display the first few rows of the DataFrame
house_df.head()
```

```python
sns.displot(house_df['Price'])
```

```python
sns.distplot(house_df['Price'])
```

```python
corrrelation = house_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corrrelation, annot=True, cmap='Blues', annot_kws={'size':8}, square=True,fmt='.1f')
# plt.show()
```
