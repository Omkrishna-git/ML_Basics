```python
x = np.linspace(0,10,20)
y = np.sin(x)
z = np.cos(x)

fid = plt.figure()
ax1 = fid.add_axes([0,0,0.5,0.5])

ax1.scatter(x,y,color='b',label='sine')
ax1.scatter(x,z,color='r',label='cosine')

plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')

z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
