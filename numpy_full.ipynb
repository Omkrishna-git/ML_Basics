```python
a = np.random.randint(1,10,(4,4))
b = np.random.randint(1,10,(4,4))

print(a)
print(b)

print("Add")
print(a+b)
print("Sub")
print(a-b)
print("Mul")
print(a*b)
print("Div")
print(a/b)

print("Add")
print(np.add(a,b))
print("Sub")
print(np.subtract(a,b))
print("Mul")
print(np.multiply(a,b))
print("Div")
print(np.divide(a,b))
```
